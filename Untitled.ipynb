```python
df.drop(columns=['customerID'],axis=1,inplace=True)
```

```python
categorical_feature=[feature for feature in df.columns if df[feature].dtypes=='O' and feature!='TotalCharges']
numerical_feature=[feature for feature in df.columns if df[feature].dtypes!='O' or feature=='TotalCharges']
```

# EDA

```python
for feature in categorical_feature:
    plt.figure(figsize=(9,5))
    sns.countplot(x=feature,data=df,hue='Churn')
    plt.show()
    
```

### Partner and Dependents:

Observation: Customers who have a partner or dependents are less likely to churn.

Explanation: Customers with family ties might exhibit higher loyalty, possibly due to family-based subscription needs or shared benefits.

### Internet Service Type:

Observation: Customers using fiber optic internet service are more likely to churn than those using DSL or no internet.

Explanation: This may reflect dissatisfaction with fiber optic pricing or service quality, or a potential association between fiber optic users and higher expectations.

### Tech Support:

Observation: Customers who do not have tech support are more likely to churn.

Explanation: Tech support likely enhances customer satisfaction by providing troubleshooting and help, reducing frustration and making customers feel valued.

### Contract Type:

Observation: Customers with month-to-month contracts are most likely to churn compared to those with one-year or two-year contracts.

Explanation: Month-to-month contracts offer greater flexibility, so customers can leave without significant penalties, making them more prone to churn.

### Paperless Billing:

Observation: Customers who have opted for paperless billing are more likely to churn than those receiving paper bills.

Explanation: This might indicate less engagement, as customers with paper bills may engage more by receiving tangible payment reminders, which could lead to a greater sense of commitment.

### Payment Method:

Observation: Customers who use electronic checks are more likely to churn than those using other payment methods, such as bank transfers, credit cards, or automatic payments.

Explanation: Electronic checks might be associated with more temporary customers, or perhaps less convenience, which could make retention challenging.
